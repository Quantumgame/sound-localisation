# Gamatone filter processing

## Imports and function definitions

```python
from brian import *
from brian.hears import *
from matplotlib import pyplot


def bpm_to_index(bpm, sample_rate, start, bars):
    seconds_per_bar = 60.0/bpm*4
    start_index = int(seconds_per_bar*sample_rate*start)
    end_index = start_index + int(sample_rate*seconds_per_bar*bars)
    return (start_index, end_index)

left120 = Sound("sounds/Recordings/playback 68 120 flat.wav")
right120 = Sound("sounds/Recordings/playback 0 120 flat.wav")
```

## Initial experiments of plotting gamatone filters

```python
a,b = bpm_to_index(100, 44100, 0, 0.25)
cfmin, cfmax, cfN = 20*Hz, 20*kHz, 100
cf = erbspace(cfmin, cfmax, cfN)
gfb_left120 = Gammatone(left120.left[a:b], cf)
gfb_right120 = Gammatone(left120.right[a:b], cf)
gfb_left_120 = Gammatone(right120.left[a:b], cf)
gfb_right_120 = Gammatone(right120.right[a:b], cf)
#ihc = FunctionFilterbank(gfb, lambda x: clip(x, 0, Inf)**(1.0/3.0))
gt_mon_left120 = gfb_left120.process()
gt_mon_right120 = gfb_right120.process()
gt_mon_left_120 = gfb_left_120.process()
gt_mon_right_120 = gfb_right_120.process()
```

```python
figure()
figsize(15,10)
imshow(gt_mon.T, aspect='auto', origin='lower left',
       extent=(0, left120.left[a:b].duration/ms,
               center_frequencies[0], center_frequencies[-1]))
pyplot.yscale('log')
title('Cochleogram')
ylabel('Frequency (Hz)')
xlabel('Time (ms)')

show()
```

# Export of gamatone 

This runs the gamatone filter on all recordings and saves the results to csv files.... takes a long time!

```python
import os
import glob
path = 'sounds/Recordings/'

def writeCSV(filename, index, array, side):
    savetxt(filename+'-'+side+'-bar_'+str(index)+'.csv',array,delimiter=",")
    
cfmin, cfmax, cfN = 20*Hz, 20*kHz, 100
cf = erbspace(cfmin, cfmax, cfN)

leftmeans = {}
rightmeans = {}

for f in glob.glob( os.path.join(path, '*.wav') ):
    curr_sound = Sound(f)
    for i in range(0,13):
        a,b = bpm_to_index(100, 44100, i, 1)
        idx = f+'-bar:'+str(i)
        
        lefttones = Gammatone(curr_sound.left[a:b], cf)
        lefttones_proc = lefttones.process()
        leftmeans[idx] = mean(lefttones_proc,0)
        writeCSV(f,i,lefttones_proc,'left')
        writeCSV(f,i,leftmeans[idx],'left_mean')
        
        righttones = Gammatone(curr_sound.right[a:b], cf)
        righttones_proc = righttones.process()
        rightmeans[idx] = mean(righttones_proc,0)
        writeCSV(f,i,righttones_proc,'right')
        writeCSV(f,i,rightmeans[idx],'right_mean')
    
    
```

#Example of Gamatone filter from the Brian Website

```python
sound = whitenoise(100*ms).ramp()
sound.level = 50*dB

nbr_center_frequencies = 50
b1 = 1.019  #factor determining the time constant of the filters
#center frequencies with a spacing following an ERB scale
center_frequencies = erbspace(100*Hz, 1000*Hz, nbr_center_frequencies)
gammatone = Gammatone(sound, center_frequencies, b=b1)

gt_mon = gammatone.process()

figure()
figsize(15,10)
imshow(gt_mon.T, aspect='auto', origin='lower left',
       extent=(0, sound.duration/ms,
               center_frequencies[0], center_frequencies[-1]))
pyplot.yscale('log')
title('Cochleogram')
ylabel('Frequency (Hz)')
xlabel('Time (ms)')

show()
```
