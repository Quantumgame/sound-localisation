```python
import scikits.audiolab as audio
import scipy
from __future__ import division

def smooth(x,window_len=11,window='hanning'):
        if x.ndim != 1:
                raise ValueError, "smooth only accepts 1 dimension arrays."
        if x.size < window_len:
                raise ValueError, "Input vector needs to be bigger than window size."
        if window_len<3:
                return x
        if not window in ['flat', 'hanning', 'hamming', 'bartlett', 'blackman']:
                raise ValueError, "Window is on of 'flat', 'hanning', 'hamming', 'bartlett', 'blackman'"
        s=numpy.r_[2*x[0]-x[window_len-1::-1],x,2*x[-1]-x[-1:-window_len:-1]]
        if window == 'flat': #moving average
                w=numpy.ones(window_len,'d')
        else:  
                w=eval('numpy.'+window+'(window_len)')
        y=numpy.convolve(w/w.sum(),s,mode='same')
        return y[window_len:-window_len+1]

def samp(start, sample_rate, end = None):        #return array values from time in seconds
    if end is None:
        return start*sample_rate
    else:
        return (start*sample_rate, end*sample_rate)

def bpm_to_index(bpm, sample_rate, start, bars):
    seconds_per_bar = 60/bpm*4
    start_index = int(seconds_per_bar*sample_rate*start)
    end_index = start_index + int(sample_rate*seconds_per_bar*bars)
    return (start_index, end_index)

def plot_frequencies(bars, audio, sample_rate):
    fig, ax = plt.subplots(1)
    fig.set_size_inches(14,7)
    for bar in bars:
        a,b = bpm_to_index(100, sample_rate, bar, 0.5)
        fft = np.fft.rfft(audio[a:b])
        freq_ave = smooth(fft, 600, 'hamming')
        ax.plot(range(0,len(freq_ave)), log(freq_ave), label="Bar "+str(bar))
        
    ax.legend(loc='upper left')
        

def setup_graph(title='', x_label='', y_label='', fig_size=None):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0], fig_size[1])
    ax = fig.add_subplot(111)
    ax.set_title(title)
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    
def plot_channels(start = 0, end = 1000):
    fig, ax = plt.subplots(1)
    fig.set_size_inches(14,7)
    ax.plot(time_array[start:end], right[start:end],label = "Right Channel", color="red", alpha = 0.5)
    ax.plot(time_array[start:end], left[start:end],label = "Left Channel", color="blue", alpha = 0.5)
    ax.set_title("Channel Plots")
    ax.set_xlabel("Time (sec)")
    ax.set_ylabel("Amplitude")
    ax.legend(loc='upper left')
    
def offset_channel(channel, offset_range = 500):
    outarray = channel[0:-offset_range]
    for i in xrange(1,offset_range):
        outarray = vstack((outarray,channel[i:i-offset_range]))
    return outarray.T
```
